```python
s = 0
i = 0
n = int(input("Enter n = "))
while i<n:
    x = int(input("Enter x = "))
    s = s+x
    i = i+1
print(s)
```

```python
# s = ""
i = 0
s = str()
n = int(input("Enter n = "))
while i<n:
    name = input("Enter name = ")
    phone = input("Enter phone = ")
    s = s + name + " " +phone + "\n"
    i+=1
print(s)
```

```python
# control statements
# break
# continue
# pass
```

```python
#break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#continue statement
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
a = "Hello World I am Python. I am the best."
for i in a:
    if i == ".":
        break
    print(i,end = "")
```

```python
a = "Hello World I am Python. I am the best."
for i in a:
    if i == ".":
        continue
    print(i,end = "")
```

```python
# pass statement
a = 3
if a >10:
    pass
```

```python
# control statements in while loop
#break
a = 0
while a<=5:
    if a ==3:
        break
    print(a)
    a = a+1
```

```python
#continue
a = 0
while a<=5:
    if a ==3:
        a = a+1
        continue
    print(a)
    a = a+1
```
