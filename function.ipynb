```python
# function with argument and no return type
def cal(x):                       #l,b,h => parameter
    l = x[0]
    b = x[1]
    h = x[2]
    v = l*b*h
    print(v,"m^3")
   

l = int(input("Enter l = "))     #global variable
b = int(input("Enter b = "))     #global variable
h = int(input("Enter h = "))     #global variable
x = [l,b,h]
cal(x)                       #l,b,h => argument
```

```python
#function with argument and return type
def cal(x,y,z):
    a = x*y
    v = x*y*z
    return a,v

l = int(input("Enter l = "))     #global variable
b = int(input("Enter b = "))     #global variable
h = int(input("Enter h = "))     #global variable 

area, volume = cal(l,b,h)
print("The area = ",area)
print("The volume ", volume)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input('Enter b = '))
    o = input("Enter + or - or * or / ")
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/":
        print(div(a,b))
    else:
        print("Invalid operator")
        
cal()
```

```python
# # WAP to create a billing system using functon
# name price quantity total = price*quantity
# add all total and get grand total
grand_total = 0
bill = ""
def cal(n):
    global grand_total, bill
    for i in range(n):
        name = input("Enter name = ")
        price = int(input("Enter price = "))
        quantity = float(input("Enter quantity = "))
        total = price * quantity
        grand_total = grand_total +total
        bill = bill + f"{name} {price} {quantity} {total}\n"
    return bill, grand_total

n = int(input("Enter n = "))
b,g = cal(n)
print(b)
print("All total = ",g)
```
